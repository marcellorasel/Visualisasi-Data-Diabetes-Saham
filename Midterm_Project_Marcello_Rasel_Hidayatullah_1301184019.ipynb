### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

Function

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

#### Code Saham HangSeng

```python
# type your code here
df_HS["Day_Perc_Change"] = df_HS["Adj Close"].pct_change()*100
df_HS.dropna(inplace = True, axis = 0)

df_HS["Trend"]= df_HS["Day_Perc_Change"].apply(lambda x:trend(x))

df_HS['Date'] = pd.to_datetime(df_HS['Date'])
df_HS_2019 = df_HS[df_HS['Date'].dt.year == 2019]
```

### Code Saham Nasdaq

```python
# type your code here
df_Nasdaq["Day_Perc_Change"] = df_Nasdaq["Adj Close"].pct_change()*100
df_Nasdaq.dropna(inplace = True, axis = 0)

df_Nasdaq["Trend"]= df_Nasdaq["Day_Perc_Change"].apply(lambda x:trend(x))

df_Nasdaq['Date'] = pd.to_datetime(df_Nasdaq['Date'])
df_Nasdaq_2019 = df_Nasdaq[df_Nasdaq['Date'].dt.year == 2019]
```

### Code Saham Nikkei

```python
# type your code here
df_Nikkei["Day_Perc_Change"] = df_Nikkei["Adj Close"].pct_change()*100
df_Nikkei.dropna(inplace = True, axis = 0)

df_Nikkei["Trend"]= df_Nikkei["Day_Perc_Change"].apply(lambda x:trend(x))

df_Nikkei['Date'] = pd.to_datetime(df_Nikkei['Date'])
df_Nikkei_2019 = df_Nikkei[df_Nikkei['Date'].dt.year == 2019]
```

Visualization Saham HangSeng

```python
# type your code here
df_HS_pie_data = df_HS_2019.groupby('Trend')
pie_label = sorted([i for i in df_HS_2019.loc[:, 'Trend'].unique()])
plt.pie(df_HS_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham HangSeng 2019', y = 1.5)
plt.show()
```

Visualization Saham Nasdaq

```python
# type your code here
df_Nasdaq_pie_data = df_Nasdaq_2019.groupby('Trend')
pie_label = sorted([i for i in df_Nasdaq_2019.loc[:, 'Trend'].unique()])
plt.pie(df_Nasdaq_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham Nasdaq 2019', y = 1.5)
plt.show()
```

Visualization Saham Nikkei

```python
# type your code here
df_Nikkei_pie_data = df_Nikkei_2019.groupby('Trend')
pie_label = sorted([i for i in df_Nikkei_2019.loc[:, 'Trend'].unique()])
plt.pie(df_Nikkei_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham Nikkei 2019', y = 1.5)
plt.show()
```

#### Penjelasan 
Untuk Saham HangSeng, trend tertinggi terjadi pada trend Slight or No Change yaitu sebesar 48.4%, diikuti dengan Slight Positive sebesar 14.6%, lalu Positive sebesar 13.0%, lalu Negative sebesar 12.2%, lalu Slight Negative sebesar 11.4% dan terendah Among top gainers yaitu sebesar 0.4%. 

Untuk Saham Nasdaq, trend tertinggi terjadi pada trend Slight or No Change yaitu sebesar 46.0%, diikuti dengan Positive sebesar 17.1%, lalu Slight Positive sebesar 16.7%, lalu Slight Negative dan Negative sama besar, yaitu sebesar 9.1%, lalu Among top losers yaitu sebesar 1.6% dan terendah Among top gainers sebesar 0.4%.

Untuk Saham Nikkei, trend tertinggi terjadi pada trend Slight or No Change yaitu sebesar 52.7%, diikuti dengan Slight Negative sebesar 14.5%, lalu Slight Positive sebesar 13.3%, lalu Positive sebesar 11.6%, lalu Negative sebesar 7.5% dan terendah Among top losers yaitu sebesar 0.4%.



### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

Set Index

```python
df_HS_2019.set_index('Date', inplace=True)
df_Nasdaq_2019.set_index('Date', inplace=True)
df_Nikkei_2019.set_index('Date', inplace=True)
```

#### Code

```python
# type your code here
# merge Adj. close values
combined_df = pd.concat([df_HS_2019["Adj Close"], df_Nasdaq_2019["Adj Close"], df_Nikkei_2019["Adj Close"]], axis=1)
combined_df.columns = ["HangSeng", "Nasdaq", "Nikkei"]

# drop null values
combined_df.dropna(inplace = True, axis = 0)

# store daily returns of all above stocks in a new dataframe 
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)

volatility = pct_chg_df[['HangSeng', 'Nasdaq', 'Nikkei']].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (18, 6))
plt.title('Volatilitas Saham HangSeng, Nasdaq dan Nikkei 2019')
plt.xlabel('Hari')
plt.ylabel('Harga')
```

#### Penjelasan 
Pada grafik volatilitas saham HangSeng, Nasdaq dan Nikkei selama tahun 2019 tersebut cenderung sama, yaitu trend Slight or No Change berada pada range harga 1-3. Untuk harga tertinggi terjadi pada saham Nasdaq yang terjadi sekitar bulan 8 2019 dan harga saham terendah terjadi pada saham Nikkei yang terjadi saat mendekati bulan 11 2019 atau bulan 10 akhir.
